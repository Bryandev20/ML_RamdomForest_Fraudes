## Modelo RamdomForest

```python
x=data.drop('Class', axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelo=RandomForestClassifier(n_estimators=300,max_depth=10,class_weight= 'balanced',random_state=42)
```

```python
modelo.fit(x_train,y_train)
```

```python
y_pred=modelo.predict(x_test)
y_precit_proba=modelo.predict_proba(x_test)[:,1]
```

## Evaluacion del modelo

```python
from sklearn.metrics import precision_score,recall_score,confusion_matrix, f1_score,roc_auc_score
precis=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
conf=confusion_matrix(y_test,y_pred)
f1=f1_score(y_test,y_pred)
au=roc_auc_score(y_test,y_precit_proba)

print('Precision =',precis)
print('Recall =',recall)
print('Confusion Matrix =',conf)
print('F1 Score =',f1)
print('auc =',au)
```
